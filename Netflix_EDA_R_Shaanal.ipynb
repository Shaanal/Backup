#### Chart - 6 & 7
##### Column: `duration`

````python
# Chart - 6&7 visualization code
````

````python
# Separate movies and TV shows
movies = nf[nf['type'] == 'Movie'].copy()
tv_shows = nf[nf['type'] == 'TV Show'].copy()

# Extract numeric duration for movies
movies['duration_min'] = movies['duration'].str.extract('(\d+)').astype(float)

# Define bins for movie durations
bins = [0, 60, 90, 120, 180, np.inf]
labels = ['<60 min', '60-90 min', '90-120 min', '120-180 min', '>180 min']
movies['duration_range'] = pd.cut(movies['duration_min'], bins=bins, labels=labels, right=False)

# Plot movie duration ranges
plt.figure(figsize=(8, 4))
movies['duration_range'].value_counts().sort_index().plot(kind='bar', color='coral')
plt.title('Distribution of Movie Durations (Binned)')
plt.xlabel('Duration Range')
plt.ylabel('Number of Movies')
plt.show()

# For TV shows, extract number of seasons
tv_shows['num_seasons'] = tv_shows['duration'].str.extract('(\d+)').astype(float)

# Plot number of seasons
plt.figure(figsize=(8, 4))
tv_shows['num_seasons'].value_counts().sort_index().plot(kind='bar', color='teal')
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of TV Shows')
plt.show()
````

##### 1. Why did you pick the specific chart?

Since both types are considerably different duration wise, it is best to analyze them separately. A bar chart with binned ranges was chosen because individual movie durations vary widely and are not meaningful when shown as unique values. A bar chart of the number of seasons was chosen because TV show durations are best represented by the number of seasons, not minutes. This chart shows the most common TV show lengths in terms of seasons.

##### 2. What is/are the insight(s) found from the chart?

The chart reveals the 90-120 min range has the highest bar for movies, it means most movies on Netflix are feature-length films. The chart shows whether Netflix has more mini-series (1 season) or long-running shows (multiple seasons). The high count for 1 Season suggests many limited series or new shows.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, these insights help Netflix understand viewer preferences for movie length. If most movies are in a narrow range, Netflix could consider adding more short or long films to attract a broader audience. A lack of variety in durations could lead to negative growth by not catering to all viewer segments. Knowing the distribution of TV show lengths helps Netflix balance its catalog. If there are too many short series, Netflix may want to invest in longer-running shows to increase user retention. Conversely, a lack of mini-series could mean missing out on viewers who prefer shorter commitments.

````python
# Extract numeric duration for movies (in minutes)
def extract_minutes(duration):
	if 'min' in duration:
		return int(duration.split(' ')[0])
	return None

# Filter only movies and extract duration in minutes
movies = nf[nf['type'] == 'Movie'].copy()
movies['duration_min'] = movies['duration'].apply(extract_minutes)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies, x='release_year', y='duration_min', alpha=0.5)
plt.title('Movie Duration (min) vs Release Year')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.tight_layout()
plt.show()

# Extract numeric duration for TV Shows (number of seasons)
def extract_seasons(duration):
	if 'Season' in duration:
		return int(duration.split(' ')[0])
	return None

# Filter only TV Shows and extract number of seasons
tv_shows = nf[nf['type'] == 'TV Show'].copy()
tv_shows['num_seasons'] = tv_shows['duration'].apply(extract_seasons)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=tv_shows, x='release_year', y='num_seasons', alpha=0.5)
plt.title('TV Show Number of Seasons vs Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Seasons')
plt.tight_layout()
plt.show()
````

````python
# Compute correlation matrix for all columns by converting categorical columns to counts/frequencies

# Prepare a DataFrame with only numeric columns
corr_df = nf[['release_year', 'year_added']].copy()

# For each categorical column, add columns with value counts (one-hot encoding for top N categories)
categorical_cols = ['type', 'rating', 'country']
for col in categorical_cols:
	top_categories = nf[col].value_counts().head(5).index  # Use top 5 to avoid too many columns
	for cat in top_categories:
		corr_df[f"{col}_{cat}"] = (nf[col] == cat).astype(int)

# For genres, use top 5 genres
for genre in top_15_genres[:5]:
	corr_df[f"genre_{genre}"] = nf['listed_in'].str.contains(genre).astype(int)

# Compute correlation matrix
corr_map = corr_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr_map, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Map (Including Encoded Categorical Features)')
plt.tight_layout()
plt.show()
````

#### Chart - 20 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
````

````python
df =nf.copy()

df['type_code'] = df['type'].astype('category').cat.codes
df['country_code'] = df['country'].astype('category').cat.codes
df['rating_code'] = df['rating'].astype('category').cat.codes
df['director_code'] = df['director'].astype('category').cat.codes
df['cast_code'] = df['cast'].astype('category').cat.codes
df['listed_in_code'] = df['listed_in'].astype('category').cat.codes

# Extract duration in minutes or seasons as numbers
def extract_duration(x):
    if 'Season' in x:
        return int(x.split()[0])
    else:
        return int(x.split()[0])

df['duration_int'] = df['duration'].apply(extract_duration)

# Select columns for correlation
correlation_cols = [
    'release_year',
    'type_code',
    'country_code',
    'rating_code',
    'duration_int',
    'director_code',
    'cast_code',
    'listed_in_code'
]

# Compute correlation matrix
corr_matrix = df[correlation_cols].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Encoded Features')
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap was chosen because it visually summarizes the relationships between multiple encoded categorical and numerical features in the Netflix dataset. This chart makes it easy to identify which features are strongly or weakly correlated, both positively and negatively.


##### 2. What is/are the insight(s) found from the chart?

- There is a very strong negative correlation between `type_code` and `duration_int` (-0.88), indicating that the type of content (Movie or TV Show) is closely linked to the duration: movies have durations in minutes, while TV shows have durations in seasons.

- `type_code` and `director_code` have a strong positive correlation (0.58), suggesting that certain directors are more associated with either movies or TV shows.
- `duration_int` is moderately negatively correlated with `rating_code` (-0.35) and `listed_in_code` (-0.33), indicating that as duration increases, the rating and genre codes tend to decrease (or vice versa).
- `release_year` has a weak positive correlation with `type_code` (0.17) and `rating_code` (0.23), suggesting that newer releases are slightly more likely to be of a certain type or rating.
- Most other correlations are weak or close to zero, indicating little linear relationship between those pairs of features.
